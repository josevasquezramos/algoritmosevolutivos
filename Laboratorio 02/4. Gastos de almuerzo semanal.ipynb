```python
import pandas as pd
```

```python
datos = {
    'Dia': ['Lunes', 'Martes', 'Miercoles', 'Jueves', 'Viernes'],
    'Gastos': [4.0, 3.5, 5.0, 4.2, 3.8]
}
```

```python
# Usamos pandas para construir el DataFrame
df = pd.DataFrame(datos)

# Visualizamos los datos
df.head()
```

```python
estadisticas = df.describe()
estadisticas
```

```python
total = df['Gastos'].sum()
promedio = df['Gastos'].mean()
```

```python
print("Gasto total de la semana:", total)
print("Gasto medio de la semana:", promedio)
```

```python
dias_sobre_promedio = df[df['Gastos'] > promedio]
print("Días en los que el gasto superó el promedio:")
dias_sobre_promedio
```
