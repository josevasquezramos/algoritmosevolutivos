```python
import pandas as pd
```

Definimos los datos

```python
datos = {
    'Estudiante': ['Rosa', 'David', 'Elena', 'Mario', 'Paula'],
    'Dias_prestamo': [7, 10, 5, 12, 3]
}
```

```python
# Usamos pandas para construir el DataFrame
df = pd.DataFrame(datos)

# Visualizamos los datos
df.head()
```

Calculamos estadisticas descriptivas

```python
estadisticas = df['Dias_prestamo'].describe()
estadisticas
```

```python
# Obtenemos el promedio
mean = estadisticas['mean']

# Obtenemos el maximo
max = estadisticas['max']
```

```python
print(f'El promedio de dias de prestamo es: {mean}')
print(f'El maximo de dias de prestamo es: {max}')
```

```python
parametro = 8
prestamo_filtro = df[df['Dias_prestamo'] > parametro]
```

```python
print(f'Los alumnos que retuvieron el libro por más de {parametro} días fueron:')
prestamo_filtro.head()
```
